```python

data = json.load(open('out/ouput.json'))
fields_df = pd.json_normalize(data,record_path=["field_values"])
# select first for each field 
fields_df = fields_df.groupby("label").agg({
    "label": "count",
    "value": "first",
})
fields_df.rename(columns={"label": "count"}, inplace=True)


fields_df.reset_index().sort_values("count").to_json("out/answers.json",orient="records",indent=2)
fields_df
```
